```python
# importações
import spacy
import nltk
nltk.download('rslp')
from spacy import displacy
from pdfminer.high_level import extract_text
```

```python
# carregando objeto nlp
nlp = spacy.load("pt_core_news_sm")
type(nlp)
# lendo PDF
text = extract_text('hp_pg1.pdf') #insira seu texto pdf
# definindo o doc que será utilizado
doc = nlp(text)
```

```python
# displacy - árvore mostrando relações do documento de texto
displacy.render(doc, style="dep")
```

```python
# tokenização - separação do texto
[token.text for token in doc]
```

```python
# debugando o tokenizador - definindo o que é token e encontrando excessões
explicacao = nlp.tokenizer.explain(text)
for token in explicacao:
    print(token[0], "\t", token[1])
```

```python
# segmentador de sentenças
for sentenca in doc.sents:
    print(sentenca.text)
```

```python
# diferença entre lemmatização e stemming
# lemma (nível léxico da palavra)
# stemming (nível morfológico da palavra)
# porém, o Spacy não trabalha a nível morfológico da palavra, portanto é necessário usar o NLTK
stemmer = nltk.stem.RSLPStemmer()
for token in doc:
    print(token.text, "|", token.lemma_, "|", stemmer.stem(token.text))
```

```python
# identificando entidades pré definidas
print(doc.ents)
```

```python
# tagging
for token in doc:
    print(token.text, "|", token.tag_, "|", spacy.explain(token.tag_))
```

```python
# relações sintáticas e dependências
for token in doc:
    print(token.text, "|", token.tag_, "|", spacy.explain(token.tag_), "|", token.dep_, "|", spacy.explain(token.dep_), "|", token.head)
displacy.render(doc, style="dep")
```

```python
# reconhecendo entidades
for entidade in doc.ents:
    print(entidade, "|", entidade.label_)
```

```python
displacy.render(doc, style="ent")
```
